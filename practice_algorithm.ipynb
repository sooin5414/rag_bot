# 우선순위 큐
우선순위 큐는 힙 기반으로 동작이고
힙 정렬은 힙을 이용한 정렬 알고리즘이다

1) Heap 자료구조 이해
- 완전 이진트리 기반 루트->왼쪽->오른쪽 순서
- 부모 <= 자식(min-heap)
- pop / push / heapify의 동작 이해
- 시간 복잡도 : 시간 복잡도: O(log N)

2) Heap을 이용한 Priority Queue(우선순위 큐)
- Heap을 그대로 내부 구현을 사용하는 자료구조
- 항상 우선 순위가 가장 높은 요소를 반환(최소 or 최대)

3) Heapsort
- 힙을 이요한 정렬 알고리즘
- 리스트 전체를 힙으로 만들고 -> 하나씩 pop해서 정렬된 리스트 생성

Heap (힙 기본 원리)
      ↓
Priority Queue (힙을 기반으로 사용)
      ↓
Heapsort (힙을 이용한 정렬 알고리즘)
