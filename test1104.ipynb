```python
import whisper
import json
import glob
import os
import subprocess

# 1. URL ë§¤í•‘ ë¨¼ì € ìƒì„±
print("ðŸ” í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ URL ë§¤í•‘ ì¤‘...\n")

PLAYLIST_URL = "https://www.youtube.com/playlist?list=PLs7TzBwHKDjLo3I1dfAOsej2S1vW1VjHG"

result = subprocess.run(
    ["yt-dlp", "--dump-json", "--skip-download", "--playlist-items", "1-30", PLAYLIST_URL],
    capture_output=True, text=True
)

video_url_map = {}
for line in result.stdout.strip().split('\n'):
    if line:
        info = json.loads(line)
        idx = info['playlist_index']
        title = info['title']
        url = info['webpage_url']
        json_filename = f"{idx:02d}_{title}.json"
        video_url_map[json_filename] = url

print(f"âœ… URL ë§¤í•‘ ì™„ë£Œ: {len(video_url_map)}ê°œ\n")

# 2. Whisper transcription
print("ðŸŽ¤ Transcription ì‹œìž‘...\n")

model = whisper.load_model("large")  # ë˜ëŠ” "small", "medium", "large"
mp3_files = sorted(glob.glob("./youtube_playlist/*.mp3"))

print(f"ðŸ“ ë°œê²¬ëœ mp3 íŒŒì¼: {len(mp3_files)}ê°œ\n")

for mp3_file in mp3_files:
    filename = os.path.basename(mp3_file)
    json_filename = mp3_file.replace('.mp3', '.json')
    
    # ì´ë¯¸ JSON ìžˆìœ¼ë©´ ìŠ¤í‚µ
    if os.path.exists(json_filename):
        print(f"â­ï¸  {filename} - ì´ë¯¸ ì¡´ìž¬")
        continue
    
    print(f"ðŸŽ¤ {filename} - ì²˜ë¦¬ ì¤‘...")
    
    # Whisper transcription
    result = model.transcribe(
        mp3_file,
        language="ko",
        verbose=False
    )
    
    # JSON ì €ìž¥
    with open(json_filename, 'w', encoding='utf-8') as f:
        json.dump(result, f, ensure_ascii=False, indent=2)
    
    print(f"âœ… {filename} - ì™„ë£Œ\n")

# 3. video_url_map ì¶œë ¥
print("\n" + "="*80)
print("ðŸ“ ë³µë¶™ìš© video_url_map:\n")
print("video_url_map = {")
for k, v in video_url_map.items():
    print(f'    "{k}": "{v}",')
print("}")
print("="*80)

print("\nðŸŽ‰ ëª¨ë“  ìž‘ì—… ì™„ë£Œ!")
```

```python
import unicodedata

def normalize(s):
    # ì „ê°ë¬¸ìž â†’ ë°˜ê°ë¬¸ìž, ê³µë°±/ë¹„ê°€ì‹œë¬¸ìž ì œê±°
    s = unicodedata.normalize("NFKC", s)
    s = s.replace("\xa0", " ").replace("  ", " ").strip()
    return s

video_url_map_norm = {normalize(k): v for k, v in video_url_map.items()}
json_files = glob.glob(os.path.join("./youtube_playlist", "*.json"))


MAX_DURATION = 30.0   # ì²­í¬ë‹¹ ìµœëŒ€ ê¸¸ì´(ì´ˆ)
OVERLAP_SEC = 5.0     # ë‹¤ìŒ ì²­í¬ì™€ ê²¹ì¹  ì‹œê°„(ì´ˆ)
MIN_TEXT_LEN = 50     # ë„ˆë¬´ ì§§ì€ ê²½ìš° ë” ë³‘í•©

all_documents = []

# =============================================
# 4. ë©”ì¸ ë£¨í”„
# =============================================
for json_file in json_files:
    video_filename = os.path.basename(json_file)
    video_filename_norm = normalize(video_filename)
    
    # URL ë§¤í•‘
    video_url = video_url_map_norm.get(video_filename_norm, "URL_ì—†ìŒ")
    if video_url == "URL_ì—†ìŒ":
        for k, v in video_url_map_norm.items():
            if os.path.splitext(video_filename_norm)[0] in k:
                video_url = v
                break

    # JSON ë¡œë“œ
    with open(json_file, "r", encoding="utf-8") as f:
        result = json.load(f)
    segments = result["segments"]

    # =============================================
    # 5. ì‹œê°„ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©
    # =============================================
    current_chunk = []
    current_start = None
    current_end = None

    for seg in segments:
        start, end, text = seg["start"], seg["end"], seg["text"]
        if current_start is None:
            current_start = start
        current_end = end
        current_chunk.append(seg)

        duration = current_end - current_start
        text_len = sum(len(s["text"]) for s in current_chunk)

        # chunk ê¸¸ì´ ì¡°ê±´
        if duration >= MAX_DURATION and text_len >= MIN_TEXT_LEN:
            doc = Document(
                page_content=" ".join(s["text"] for s in current_chunk),
                metadata={
                    "video_url": video_url,
                    "start_time": current_start,
                    "end_time": current_end,
                },
            )
            all_documents.append(doc)

            # overlap ì²˜ë¦¬ (ì•žë¶€ë¶„ ì¡°ê¸ˆ ë‚¨ê¸°ê¸°)
            overlap_start = max(current_end - OVERLAP_SEC, current_start)
            current_chunk = [s for s in current_chunk if s["end"] > overlap_start]
            current_start = current_chunk[0]["start"] if current_chunk else None

    # ë§ˆì§€ë§‰ ë‚¨ì€ chunk ì¶”ê°€
    if current_chunk:
        doc = Document(
            page_content=" ".join(s["text"] for s in current_chunk),
            metadata={
                "video_url": video_url,
                "start_time": current_chunk[0]["start"],
                "end_time": current_chunk[-1]["end"],
            },
        )
        all_documents.append(doc)

print(f"âœ… ë¬¸ì„œ {len(all_documents)}ê°œ ì¤€ë¹„ ì™„ë£Œ! (ì‹œê°„ ë‹¨ìœ„ ë³‘í•©, overlap={OVERLAP_SEC}s)")
```
